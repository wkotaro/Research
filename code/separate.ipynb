```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../image/sample.jpg")
plt.imshow(img[:,:,::-1])
```

```python
img2 = img.copy()
i = 30
while(i < 300):
    img2 = cv2.line(img2,(i ,0),(i ,300),(255,0, 0),5)
    i += 60
plt.imshow(img2[:,:,::-1])
```

```python
img3 = img2.copy()
i = 30
while(i < 300):
    img3 = cv2.line(img3,(0 ,i),(300 ,i),(255,0, 0),5)
    i += 60
plt.imshow(img3[:,:,::-1])
```
